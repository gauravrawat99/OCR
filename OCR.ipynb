```python
#inverting
invert_img=cv2.bitwise_not(img)
cv2.imwrite("temp/ivimg.jpg",invert_img)
```

```python
image = cv2.imread("temp/ivimg.jpg")
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

```python
#binarization
greyscale =cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imwrite("temp/grey.jpg",greyscale)
```

```python
thresh,im_bw = cv2.threshold(greyscale,0,0,cv2.THRESH_BINARY)
cv2.imwrite("temp/im_bw.jpg",greyscale)
```

```python
image = cv2.imread("temp/im_bw.jpg")
from google.colab.patches import cv2_imshow
cv2_imshow(image) 
```

```python
#noise removal
kernel = np.ones((1,1),np.uint8)
image = cv2.dilate(image,kernel,iterations=1)
kernel = np.ones((1,1),np.uint8)
image = cv2.erode(image,kernel,iterations=1)
image = cv2.morphologyEx(image,cv2.MORPH_CLOSE,kernel)
image = cv2.medianBlur(image,3)
cv2.imwrite("temp/noise.jpg",image)
image = cv2.imread("temp/noise.jpg")
from google.colab.patches import cv2_imshow
cv2_imshow(image) 
```

```python

out = pytesseract.image_to_string(image)
print(out)
```
